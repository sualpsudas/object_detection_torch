```python
import os
from random import choice
import shutil
```

```python
os.listdir(os.getcwd()+"\\banana")
```

```python
c=[]
for i in os.listdir(os.getcwd()+"\\banana"):
    if i.endswith(".jpg"):
        c.append(i.split(".")[0])
```

```python
train=[]
val=[]
0.8*len(c)
```

```python
c_=0
while c_<161:
    num=choice(c)
    train.append(num)
    c.remove(num)
    c_+=1
val=c.copy()
```

```python
len(train)+len(val)
```

```python
for i in train:
    shutil.move(os.getcwd()+"\\banana\\"+str(i)+".jpg",os.getcwd()+"\\data\\images\\train\\"+str(i)+".jpg")
    shutil.move(os.getcwd()+"\\banana\\"+str(i)+".txt",os.getcwd()+"\\data\\labels\\train\\"+str(i)+".txt")
```

```python
for i in val:
    shutil.move(os.getcwd()+"\\banana\\"+str(i)+".jpg",os.getcwd()+"\\data\\images\\val\\"+str(i)+".jpg")
    shutil.move(os.getcwd()+"\\banana\\"+str(i)+".txt",os.getcwd()+"\\data\\labels\\val\\"+str(i)+".txt")
```
